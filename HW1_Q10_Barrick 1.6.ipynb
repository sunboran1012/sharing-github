```python
//Boran Sun
//bsun24
!conda install -c anaconda seaborn -y
import numpy as np 
import random 
import matplotlib.pyplot as plt 
import seaborn as sns 
np.random.seed(42)
n=10 
p=0.5 
trials=100 
def bionomial(trials,n,p): 
    heads=[] 
    for i in range(trials): 
        flips=[np.random.random() for i in range (n)] 
        heads.append(len([i for i in flips if i >=0.50])) 
    return heads 
heads=bionomial(trials,n,p) 
fig,ax=plt.subplots() 
ax = sns.distplot(heads, bins=11) 
ax.set_xlabel("Number of Heads") 
ax.set_ylabel("Frequency")
plt.xticks(np.arange(0,11,1))
import statistics
statistics.stdev(heads)

#Reference: https://cmdlinetips.com/2018/12/simulating-coin-toss-experiment-with-binomial-random-numbers-using-numpy/
```
